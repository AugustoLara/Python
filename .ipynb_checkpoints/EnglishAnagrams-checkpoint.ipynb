## Data Cleaning

```python
'Aaron\n'.strip()
```

```python
wordclean = sorted(list(set(([word.strip().lower() for word in wordlist]))))
# palabrasclean = [x.strip().lower() for x in listapalabras]
```

```python
wordclean[:10]
```

## Word signatures

```python
sorted('lives')
```

```python
sorted('lives') == sorted('elvis')
```

```python
sorted('love') == sorted('hate')
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
signature('lives')
```

```python
'/'.join(['a','b','c'])
```

```python
def anagram(word):
    return [w for w in wordclean if signature(w) == signature(word)]
```

```python
anagram('dictionary')
```

```python
%timeit anagram('dictionary')
```

## Creating a dictionary of words by signature

```python
import collections
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(word):
    return words_bysig[signature(word)]
```

```python
anagram_fast('dictionary')
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
len(anagrams_all)
```

```python
%timeit anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
anagrams_all
```

## Counting words by signature -1 (remove original word)
Columns are: Key=Signature Values=[Words by Signature]

```python
anagrams_byword = {index: [len(anagrams_all[index]) - 1]  for index in anagrams_all}
anagrams_byword
```

## Calculating length of signature
Columns are: Key=Signature Values=[Words by Signature., length of signature]

```python
for word in anagrams_all:
    anagrams_byword[word].append(len(word))
```

```python
anagrams_byword
```

## Calculating range of signatures

```python
word_counts=[]
for word in anagrams_byword:
    word_counts.append(anagrams_byword[word][1])
word_counts = set(word_counts)
word_counts = list(word_counts)
word_counts
```

```python
anagrams_byword['abb'][1] == word_counts[1]
```

## Adding words by range of signatures

```python

results = []
for key in range(21):
    x= 0
    for word in anagrams_byword:
        if anagrams_byword[word][1] == word_counts[key]:
            x = anagrams_byword[word][0] + x
    results.append((word_counts[key],x/2))

results
```

```python
pp.plot(results)
pp.show()
```
